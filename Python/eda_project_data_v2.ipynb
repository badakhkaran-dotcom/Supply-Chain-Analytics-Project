# KPI

### KPI-Using MYSQL

```python
query="""
SELECT 
    SUM(order_qty)AS total_order_qty,
    SUM(total_revenue)AS total_revenue,
    AVG(defect_rate)AS avg_defect_rate,
    AVG(lead_time_days)AS avg_lead_time,
    COUNT(DISTINCT product_type)AS total_product,
    COUNT(DISTINCT sku_id)AS total_sku_id,
    AVG(total_revenue)AS avg_revenue,
    AVG(order_qty)AS avg_order_qty,
    MAX(order_qty)AS maximum_order_qty,
    MIN(order_qty)AS min_order_qty,
    SUM(shipping_cost)AS total_shipping_cost,
    COUNT(DISTINCT supplier_name)AS total_supplier,
    SUM(manufacturing_cost)AS total_manufacturing_cost,
    COUNT(DISTINCT inspection_results)AS unique_QA_results,
    MAX(defect_rate)AS max_defect_rate,
    SUM(profit)AS total_profit,
    AVG(shipping_efficiency)AS avg_shipping_efficiency,
    COUNT(date)AS total_transaction
FROM supply_chain_data;

"""
df=pd.read_sql(query,engine)
print(df)
```

```python
# Supplier Wise Revenue
pd.set_option('display.float_format','{:.2f}'.format) # “Show float numbers in normal decimal format with 2 decimal places”

query="""
SELECT supplier_name,SUM(total_revenue)AS total_Revenue
FROM supply_chain_data
GROUP BY supplier_name
ORDER BY total_revenue DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
print("Supplier Wise Revenue")
supplier_wise_revenue=(
    supply_chain_data_df
    .groupby("supplier_name")["total_revenue"]
    .sum()
    .reset_index()
    .sort_values(by="total_revenue",ascending=False)
)
print(supplier_wise_revenue)
```

```python
# Order Volume Across Suppliers

query="""
SELECT 
    supplier_name,
    SUM(order_qty)AS total_order
FROM supply_chain_data
GROUP BY supplier_name
ORDER BY total_order DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python

suppliers_order_volume=(
    supply_chain_data_df
    .groupby("supplier_name")["order_qty"]
    .sum()
    .reset_index()
    .sort_values(by="order_qty",ascending=False)
)
print(suppliers_order_volume)
```

```python
# Product Category Distribution Per Supplier By order

query="""
SELECT product_type,supplier_name,SUM(order_qty)AS total_order
FROM supply_chain_data
GROUP BY product_type,supplier_name
ORDER BY product_type, total_order DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
product_category_distribution=(
    supply_chain_data_df
    .groupby(["product_type","supplier_name"])["order_qty"]
    .sum()
    .reset_index()
    .sort_values(by=["product_type","order_qty"],ascending=[True,False])
)
print(product_category_distribution)
```

```python
# Inspection Status Across supplier Qty wise
query="""
SELECT supplier_name,inspection_results,SUM(order_qty)AS total_orders
FROM supply_chain_data
GROUP BY supplier_name,inspection_results
ORDER BY supplier_name,total_orders DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
inspection_status_qty_wise=(
    supply_chain_data_df
    .groupby(["supplier_name","inspection_results"])["order_qty"]
    .sum()
    .reset_index()
    .sort_values(by=["supplier_name","order_qty"],ascending=[True,False])
)
print(inspection_status_qty_wise)
```

```python
# Defect Rate Across Supplier
query="""
SELECT supplier_name,AVG(defect_rate)AS avg_defect_rate
FROM supply_chain_data
GROUP BY supplier_name
ORDER BY avg_defect_rate DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
avg_defect_rate_across_supplier=(
    supply_chain_data_df
    .groupby("supplier_name")["defect_rate"]
    .mean()
    .reset_index()
    .sort_values(by="defect_rate",ascending=False)
)
print(avg_defect_rate_across_supplier)
```

```python
# Avg Lead Time Across Supplier
query="""
SELECT supplier_name,AVG(lead_time_days)AS avg_lead_time
FROM supply_chain_data
GROUP BY supplier_name
ORDER BY avg_lead_time DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
avg_lead_time=(
    supply_chain_data_df
    .groupby("supplier_name")["lead_time_days"]
    .mean()
    .reset_index()
    .sort_values(by="lead_time_days",ascending=False)
)
print(avg_lead_time)
```

```python
# sku Coverage across product Categories and suppliers
query="""
SELECT supplier_name,product_type,COUNT(DISTINCT(sku_id))AS count_distinct_sku_id
FROM supply_chain_data
GROUP BY supplier_name,product_type
ORDER BY supplier_name,count_distinct_sku_id DESC;
"""
df=pd.read_sql(query,engine)
print(df)
```

```python
sku_id_coverage_product_categories_suppliers=(
    supply_chain_data_df
    .groupby(["supplier_name","product_type"])["sku_id"]
    .nunique()
    .reset_index()
    .sort_values(by=["supplier_name","sku_id"],ascending=[True,False])
)
print(sku_id_coverage_product_categories_suppliers)
```
